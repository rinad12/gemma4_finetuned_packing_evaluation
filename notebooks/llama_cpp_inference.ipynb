# Gemma 4 Fine-tuned GGUF — llama-cpp-python inference

Model: `rinad12/gemma_4_fine_tuned_packing_gguf` on HuggingFace

```python
import os
from huggingface_hub import list_repo_files
from dotenv import load_dotenv

load_dotenv()

REPO_ID = "rinad12/gemma_4_fine_tuned_packing_gguf"
HF_TOKEN = os.getenv("HUGGINGFACEHUB_API_TOKEN")

files = list(list_repo_files(REPO_ID, token=HF_TOKEN))
print("Files in repo:")
for f in files:
    print(" ", f)
```

```python
from huggingface_hub import hf_hub_download

# Q4_K_M = main language model; F16-mmproj = vision projection file (not for text inference)
# Switch to gemma-4-e2b.Q4_K_M.gguf for the smaller 2B variant
GGUF_FILENAME = "gemma-4-e4b.Q4_K_M.gguf"
print("Downloading:", GGUF_FILENAME)

model_path = hf_hub_download(
    repo_id=REPO_ID,
    filename=GGUF_FILENAME,
)
print("Model cached at:", model_path)
```

```python
from llama_cpp import Llama

llm = Llama(
    model_path=model_path,
    n_ctx=2048,
    n_gpu_layers=0,  # offload all layers to GPU; set 0 for CPU-only
    verbose=False,
)
print("Model loaded.")
```

```python
# Basic completion — Gemma chat template
PROMPT = """<start_of_turn>user
Create a packing list for a 7-day beach vacation in Thailand.
<end_of_turn>
<start_of_turn>model
"""

output = llm(
    PROMPT,
    max_tokens=3000,
    temperature=0.7,
    top_p=0.9,
    stop=["<end_of_turn>"],
)

print(output["choices"][0]["text"])
```

```python
# Chat-style API (higher-level wrapper)
response = llm.create_chat_completion(
    messages=[
        {"role": "user", "content": "Create a packing list for a 3-day ski trip to the Alps."}
    ],
    max_tokens=3000,
    temperature=0.7,
)

print(response["choices"][0]["message"]["content"])
```
